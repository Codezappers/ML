```python
import numpy as np 
import matplotlib.pyplot as plt

# Data points
x = np.array([1,2,3,4,5])
y = np.array([1,2,1.3,3.75,2.25])

# Compute the bets fit line
A = np.vstack([x, np.ones(len(x))]).T
m ,b = np.linalg.lstsq( A,y, rcond = None)[0]

# Plotting
plot = plt.plot(x, y, 'o', label='Data Points')
plot = plt.plot(x, m*x + b, 'r' ,label='Best Fit Line')
plt.legend()
plt.grid()
plt.show()
```
