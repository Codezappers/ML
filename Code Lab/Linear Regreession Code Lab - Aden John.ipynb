```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data points 
np.random.seed(0)   

# rand() gives you uniformity distributed pseduo-random numbers
x = np.random.rand(50) * 10

print("x: ", x)

#normal distribution = bell like curve
y = 2 * x + 1 + np.random.randn(50) * 2
print("y: ", y)

# Use Numpy's polyfit method to perform linear regression on  given data(x,y)
# third argument which is set to '1' indicating that we want to fit a polynomial in the degree 1 

slope , intercept = np.polyfit(x, y, 1)

# Plot data points and regression line
plt.scatter(x, y, label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Regression line')
plt.ylabel('Y')
plt.xlabel('X')
plt.title('Linear Regression')
plt.legend()  
plt.grid()
plt.show()  

print("Slope: ", slope)
print("Intercept: ", intercept)    
```

```python
# Generate a new x-value for prediction
new_x = 7

m = slope
c = intercept

# Predict the corresponding y-value using the best fit line equation (y = mx + c)
predicted_y = m * new_x + c

print("For x =", new_x, "the predicted y-value is: ", predicted_y)

# Plot the original data points and the best fit line
plt.scatter(x,y, label='Data points')
plt.plot(x, m*x + c, color='red', label='Best Fit line')

# PLot the predicted point
plt.scatter(new_x, predicted_y, color='green', label='Predicted Point')
plt.ylabel('Y')
plt.xlabel('X')
plt.title('Linear Regression')
plt.legend()
plt.grid()
plt.show()

print("Slope: ", slope)
print("Intercept: ", intercept)    
```
