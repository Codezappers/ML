```python
import numpy as np
import matplotlib.pyplot as plt

# Set boundaries for random quantities
min_quantity = 50
max_quantity = 300

np.random.seed(40)

# Generate random quantities for weeks 1 and 2
week_1_quantaties = np.random.randint(min_quantity, max_quantity, size=3)
week_2_quantaties = week_1_quantaties + np.random.randint(-20, 20, size=3)

# Product names to the labels
products = ['T-shirts', 'Jeans', 'Hats']

week_1_array = np.array(week_1_quantaties) # Array of week 1 quantities
week_2_array = np.array(week_2_quantaties) # Array of week 2 quantities

# Print properties of the arrays
print('Week 1 Quantities:', week_1_array)
print("Shape of week 1 array:", week_1_array.shape)

print('Week 2 Quantities:', week_2_array)
print("Shape of week 2 array:", week_2_array.shape)

# Calculate the change in quantities
quantity_change = week_2_array - week_1_array
print("\nChange in quantities:", quantity_change)

# Prices for the products
prices = np.array([10, 15, 20]) # Prices for T-shirts, Jeans, and Hats

inventory_value_change = np.dot(quantity_change, prices)
print("\nInventory Value Change:", inventory_value_change)

# Concatenate the data into a 2D array
inventory_data = np.concatenate(([week_1_array],[week_2_array]))
print("\nInventory Data:", inventory_data)
print("Shape of inventory data:", inventory_data.shape)
print("Number of dimensions:", inventory_data.ndim)

# Plot the data
plt.plot(products, week_1_array, marker='o', label='Week 1', color='blue')
plt.plot(products, week_2_array, marker='o', label='Week 2', color='green')

# Adding labels and titles 
plt.xlabel('Products')
plt.ylabel('Quantities')
plt.title('Inventory Levels for Week 1 and Week 2')
plt.legend()

# Show plot
plt.grid()
plt.show()
```
